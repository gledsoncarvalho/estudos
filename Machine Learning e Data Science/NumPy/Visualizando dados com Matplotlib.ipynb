```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

# MATPLOTLIB = BIBLIOTECA PARA PLOTAR GRAFICOS
```

```python
salario = np.array([100,200,500,400,150])
salario
```

```python
plt.plot(salario)
```

```python
plt.plot(salario, c='black') # color = c
```

```python
plt.plot(salario, c='red', ls='--')
# c => muda a cor da linah
# ls = tipo da linha
```

```python
plt.plot(salario, c='gray', ls='--', marker='^')
# marker = marcador -- s = quadrado; o = circulo; ^ = triangulo
```

```python
plt.plot(salario, c='red', ls='--', marker='s', ms=8) # ms = tamanho do marcador
plt.show()
```
