```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
salarios_gued = np.array([100,200,400,300])
salarios_bia = np.array([50, 300, 500, 450])
```

```python
plt.plot(salarios_gued, c='black', ls='--', marker='s', ms='8', label='Salarios Gued')
plt.plot(salarios_bia, c='red', ls='--', marker='s', ms='8', label='Salarios Bia')
plt.show()
```

```python
# colocando legenda
plt.plot(salarios_gued, c='black', ls='--', marker='s', ms='8', label='Salários Gledson')
plt.plot(salarios_bia, c='red', ls='--', marker='^', ms='8', label='Salários Bia')
plt.legend()
plt.show()
```

```python
# colocando legenda
plt.plot(salarios_gued, c='black', ls='--', marker='s', ms='8', label='Salários Gledson')
plt.plot(salarios_bia, c='red', ls='--', marker='^', ms='8', label='Salários Bia')
plt.legend(loc='upper right')
plt.show()
```

```python
# colocando mais uma linha

salarios_joao = [200, 150, 300, 700]
plt.plot(salarios_gued, c='black', ls='--', marker='s', ms='8', label='Salários Gledson')
plt.plot(salarios_bia, c='red', ls='--', marker='^', ms='8', label='Salários Bia')
plt.plot(salarios_joao, c='green', ls='--', marker='o', ms='8', label='Salários João')

plt.legend(loc='upper left')
plt.show()
```
